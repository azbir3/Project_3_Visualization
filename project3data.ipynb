```python
def getmonth(month):
    
    month = str(month).split('-')[1]
    month=month[:2]
    
    if month == '01':
        month='January'
    elif month == '02':
        month='February'
    elif month == '03':
        month='March'
    elif month == '04':
        month='April'
    elif month == '05':
        month='May'
    elif month == '06':
        month='June'
    elif month == '07':
        month='July'
    elif month == '08':
        month='August'
    elif month == '09':
        month='September'
    elif month == '10':
        month='October'
    elif month == '11':
        month='November'
    elif month == '12':
        month='December'
    
    return month
    
```

```python
def gethour(time):
    
    time = int(time.split(':')[0])
    
    if time < 13 and time > 0:
        time = str(time) + " AM"
    
    if time ==13:
        time = "1 PM"
    
    elif time ==14:
        time = "2 PM"
    
    elif time ==15:
        time = "3 PM"
    
    elif time ==16:
        time = "4 PM"
    
    elif time ==17:
        time = "5 PM"
    
    elif time ==18:
        time = "6 PM"
    
    elif time ==19:
        time = "7 PM"
    
    elif time ==20:
        time = "8 PM"
    
    elif time ==21:
        time = "9 PM"
    
    elif time ==22:
        time = "10 PM"
    
    elif time ==23:
        time = "11 PM"
    
    elif time ==24:
        time = "12 PM"
       
    elif time ==0:
        time = "12 AM"
    
    
    return time
```

```python
df['HOUR'] = df['ACCIDENT TIME'].apply(gethour)
```
