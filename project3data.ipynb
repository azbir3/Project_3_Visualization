```python
def getmonth(month):
    
    month = str(month).split('-')[1]
    month=month[1]
    
    if month == '1':
        month='January'
    elif month == '2':
        month='February'
    elif month == '3':
        month='March'
    elif month == '4':
        month='April'
    elif month == '5':
        month='May'
    elif month == '6':
        month='June'
    elif month == '7':
        month='July'
    elif month == '8':
        month='August'
    elif month == '9':
        month='September'
    elif month == '10':
        month='October'
    elif month == '11':
        month='November'
    elif month == '12':
        month='December'
    
    return month
    
```

```python
df['MONTH']= df['ACCIDENT DATE'].apply(getmonth)
```

```python
df
```

```python
def gethour(time):
    
    time = int(time.split(':')[0])
    
    if time < 13 and time > 0:
        time = str(time) + " AM"
    
    if time ==13:
        time = "1 PM"
    
    elif time ==14:
        time = "2 PM"
    
    elif time ==15:
        time = "3 PM"
    
    elif time ==16:
        time = "4 PM"
    
    elif time ==17:
        time = "5 PM"
    
    elif time ==18:
        time = "6 PM"
    
    elif time ==19:
        time = "7 PM"
    
    elif time ==20:
        time = "8 PM"
    
    elif time ==21:
        time = "9 PM"
    
    elif time ==22:
        time = "10 PM"
    
    elif time ==23:
        time = "11 PM"
    
    elif time ==24:
        time = "12 PM"
       
    elif time ==0:
        time = "12 AM"
    
    
    return time
```

```python
df['HOUR'] = df['ACCIDENT TIME'].apply(gethour)
```

```python
if df['total_injuries'].iloc[0]>=1 and df['motorists'].iloc[0]>=1:
    print(True)
```

```python
df['injury_type'] = np.nan
```

```python
for i in range(len(df)):
    if df['pedestrians'].iloc[i]>=1 and df['cyclists'].iloc[i]>=1:
         df['injury_type'].iloc[i]='Mixed'
    elif df['pedestrians'].iloc[i]>=1 and df['motorists'].iloc[i]>=1:
         df['injury_type'].iloc[i]='Mixed'
    elif df['cyclists'].iloc[i]>=1 and df['motorists'].iloc[i]>=1:
         df['injury_type'].iloc[i]='Mixed'

    elif df['pedestrians'].iloc[i]>=1:
        df['injury_type'].iloc[i]='Pedestrian'
    elif df['cyclists'].iloc[i]>=1:
        df['injury_type'].iloc[i]='Cyclist'
    elif df['motorists'].iloc[i]>=1:
        df['injury_type'].iloc[i]='Motorist'

    else:
        df['injury_type'].iloc[i]='No injury'

    
```
